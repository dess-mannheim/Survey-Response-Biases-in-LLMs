# Questionnaire Validation
This chapter validates the extraction and perturbation methods applied in 01a_questionnaire_preprocessing. The goal is to have a valid and correct subset of the questionnaire including correct q&a phrasing as well as correct perturbed q/a pairs. 

```python
import pandas as pd
import matplotlib.pyplot as plt

df_validation = pd.read_excel(io="../assets/20240911_final_questionnaire_validation.xlsx", header=0)
```

## Original Q & A 
Here I check whether the transformation and extraction of q&a pairs from pdf to the dataframe has worked without flaws. 

## Synonym Perturbations 
Here I check whether the synonym perturbation of q&a pairs from the original q&a pairs has worked without flaws. Out of all questions a random subset of currently 50 q/a synonym perturbations were labelled. 

```python
flawless_synonyms = df_validation["flawless_synonym"].value_counts(dropna=True)
flawless_synonyms= flawless_synonyms/sum(flawless_synonyms)
```

```python
# Create a bar plot based on the 'Values' column
colors = ["#5A9367", "#D38D5F"]
plt.bar(["Flawless", "Erroneous"], flawless_synonyms.values,  color=colors[:len(flawless_synonyms)])

# Labeling the plot
plt.xlabel('')
plt.ylabel('Relative Frequencies')
plt.title('How many synonymizations were successful?')

# Display the plot
plt.savefig("../thesis/figures/synonymization_validation.pdf")
plt.show()
```
