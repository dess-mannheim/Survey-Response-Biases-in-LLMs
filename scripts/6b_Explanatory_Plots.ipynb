```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data for two different distributions
data1 = np.random.normal(loc=1, scale=0.2, size=1000)  # Mean=5, std=1
data2 = np.random.normal(loc=4, scale=0.2, size=1000)  # Mean=8, std=2

# Compute means of both distributions
mean_y1 = np.mean(data1)
mean_y2 = np.mean(data2)

# Plot the histograms
plt.figure(figsize=(8, 5))
plt.hist(data1, bins=11, alpha=0.2, color='#D38D5F') # , label='Original Responses')
plt.hist(data2, bins=11, alpha=0.2, color="#5A9367") # , label='Perturbed Responses')

# Get current x-axis limits
x_min, x_max = plt.xlim()

# Convert x-coordinates to proportions
plt.axhline(y=200, xmin=0.09, xmax=0.55, color='#D38D5F', linestyle='dotted', label = r"$D_{original} = \left| C - \mu_{original} \right|$")
plt.axhline(y=150, xmin=0.365, xmax=0.545, color="#5A9367", linestyle='dotted', label = r"$D_{perturbed} = \left| C - \mu_{pertubed}\right|$")

plt.text(x=2.5,y=205, s=r"$D_{original}$")
plt.text(x=4.5,y=155, s=r"$D_{perturbed}$")
# Add vertical dashed lines for the means
plt.axvline(x=mean_y1, color='#D38D5F', linestyle='dashed')
plt.axvline(x=mean_y2, color="#5A9367", linestyle='dashed')
plt.axvline(x=6, color='gray', linestyle='dashed')
# Labels and legend
plt.xlabel("Response Labels\n (Scale 1-11)")
plt.ylabel("Frequency")
plt.title("Computation of Difference in Distance")
plt.xticks(np.arange(12))
plt.legend()
plt.tick_params(
    axis='y',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    left=True,         # ticks along the top edge are off
    labelleft=False) # labels along the bottom edge are off
plt.savefig("../thesis/figures/explanatory_middle_shift_calculation.pdf")
plt.savefig("../thesis/figures/explanatory_middle_shift_calculation.png", transparent = True)
plt.show()
```
